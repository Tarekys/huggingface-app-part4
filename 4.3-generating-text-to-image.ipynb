```python
def generate_image(pipe, prompt, num_inference_steps=50, guidance_scale=7.5):
    """
    - pipe: النموذج المستخدم 
    - prompt: النص المطلوب تحويله إلى صورة.
    - num_inference_steps: عدد خطوات الاستدلال.
    - guidance_scale: تحكم في مدى التزام النموذج بالنص.(الابداع الذاتي)
    """
    print(f"Generating image for: {prompt}")
    
    image = pipe(prompt, num_inference_steps=num_inference_steps, guidance_scale=guidance_scale).images[0]
    return image
```

```python
prompt = "A mystical dragon flying over an ancient castle at sunset"
img_v15 = generate_image(pipe, prompt)
img_v15.save("image_v15.png")
img_v15.show()
```
